# Data Exploration/ waste_types

```python
waste_types = pathlib.Path('/home/loupy/code/Loupy2023/get_it_right/raw_data/waste_types') 
garb_cardboard = list(waste_types.glob('cardboard/*'))
garb_glass = list(waste_types.glob('glass/*'))
garb_paper = list(waste_types.glob('paper/*'))
garb_metal = list(waste_types.glob('metal/*'))
garb_plastic = list(waste_types.glob('plastic/*'))
garb_trash = list(waste_types.glob('trash/*'))

img = Image.open(str(garb_cardboard[100]))
img_height, img_width =  img.size
batch_size = 32
```

```python
import os
import matplotlib.pyplot as plt
# Set the path to your dataset
data_path = '/home/loupy/code/Loupy2023/get_it_right/raw_data/waste_types'
# Create an empty dictionary to store the counts for each class
class_counts = {}
# Iterate through each subdirectory in the dataset folder
for subdir in os.listdir(data_path):
    subdir_path = os.path.join(data_path, subdir)
    # Count the number of files in the subdirectory
    files_count = len(os.listdir(subdir_path))
    # Add the count to the dictionary using the subdirectory name as the key
    class_counts[subdir] = files_count
# Plot the class distribution as a bar chart
plt.bar(class_counts.keys(), class_counts.values())
plt.xlabel('Class')
plt.ylabel('Number of Images')
plt.title('Class Distribution')
plt.show()
```
